```python
columns_with_strings = df.columns[df.dtypes == 'object'].tolist()
# Elimina las columnas con valores string
df2 = df.drop(columns=columns_with_strings)
df2 = df2.dropna()
```

```python
df2.head()
```

```python
Y = df2.values
Y.shape
```

```python
sc = StandardScaler()
sc.fit(Y)
Y_scaled = sc.transform(Y)
```

```python
cov_m = np.cov(Y_scaled.T)
cov_m.shape
```

```python
eigenvalues, eigenvectors = np.linalg.eig(cov_m)
```

```python
eigenvalues
```

```python
eigenvalues.shape
```

```python
eigenvectors.shape
```

```python
eigenvectors = eigenvectors.T
```

```python
eigenvectors[0, :]
```

```python
eigenvalues = np.cumsum(eigenvalues)
eigenvalues /= eigenvalues.max()
eigenvalues
```

```python
p = eigenvectors [0:10, :]
p.shape
```

```python
data_new = np.dot(p, Y_scaled.T)
data_new.shape
```

```python
data_new = data_new.T
data_new.shape
```
